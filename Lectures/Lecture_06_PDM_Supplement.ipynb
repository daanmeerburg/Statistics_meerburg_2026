### Jackknife bias derivation

Let $\alpha_N$ denote a statistic computed from a sample of size $N$, estimating a target
quantity $\alpha$. For a broad class of smooth statistics, the finite-sample expectation
admits an expansion of the form

$$
\mathbb{E}[\alpha_N] = \alpha + \frac{b}{N} + o\!\left(\frac{1}{N}\right),
$$

where $b$ is a constant depending on the statistic and the underlying distribution.

Define $\alpha_i^*$ as the same statistic computed on the leave-one-out sample of size
$N-1$, obtained by removing observation $i$. Its expectation satisfies

$$
\mathbb{E}[\alpha_i^*]
= \alpha + \frac{b}{N-1} + o\!\left(\frac{1}{N}\right).
$$

Averaging over all leave-one-out samples yields

$$
\bar{\alpha}^* = \frac{1}{N}\sum_{i=1}^N \alpha_i^*,
\qquad
\mathbb{E}[\bar{\alpha}^*]
= \alpha + \frac{b}{N-1} + o\!\left(\frac{1}{N}\right).
$$

Consider the difference between the leave-one-out mean and the full-sample statistic:

$$
\mathbb{E}[\bar{\alpha}^* - \alpha_N]
= \left(\alpha + \frac{b}{N-1}\right)
- \left(\alpha + \frac{b}{N}\right)
+ o\!\left(\frac{1}{N}\right).
$$

The leading terms simplify to

$$
\mathbb{E}[\bar{\alpha}^* - \alpha_N]
= b\left(\frac{1}{N-1} - \frac{1}{N}\right)
+ o\!\left(\frac{1}{N}\right)
= \frac{b}{N(N-1)} + o\!\left(\frac{1}{N}\right).
$$

Multiplying both sides by $(N-1)$ gives

$$
\mathbb{E}\big[(N-1)(\bar{\alpha}^* - \alpha_N)\big]
= \frac{b}{N} + o\!\left(\frac{1}{N}\right).
$$

The term $b/N$ is exactly the leading-order bias of $\alpha_N$. Therefore, the jackknife
estimator of the bias is

$$
\widehat{\mathrm{bias}}_{\mathrm{jack}}
= (N-1)(\bar{\alpha}^* - \alpha_N).
$$

Subtracting this estimate from $\alpha_N$ yields the bias-corrected jackknife statistic

$$
\alpha^J
= \alpha_N - \widehat{\mathrm{bias}}_{\mathrm{jack}}
= \alpha_N - (N-1)(\bar{\alpha}^* - \alpha_N)
= N\alpha_N - (N-1)\bar{\alpha}^*.
$$

This construction removes the $O(1/N)$ bias term, leaving a residual bias of smaller order.


### Jackknife bias correction and standard error

Let $\alpha_N$ denote the statistic computed on the full sample of size $N$, and let
$\alpha_i^*$ denote the leave-one-out statistic computed on the sample of size $N-1$ with
observation $i$ removed. Define the leave-one-out mean

$$
\bar{\alpha}^*=\frac{1}{N}\sum_{i=1}^N \alpha_i^*.
$$

#### Bias-corrected jackknife estimate

The jackknife bias estimate is

$$
\widehat{\mathrm{bias}}_{\mathrm{jack}}=(N-1)\big(\bar{\alpha}^*-\alpha_N\big),
$$

so the bias-corrected (jackknife) statistic is

$$
\alpha^J=\alpha_N-\widehat{\mathrm{bias}}_{\mathrm{jack}}
= \alpha_N-(N-1)(\bar{\alpha}^*-\alpha_N)
= N\alpha_N-(N-1)\bar{\alpha}^*.
$$

Equivalently, defining the pseudo-values

$$
p_i = N\alpha_N - (N-1)\alpha_i^*,
$$

we have

$$
\alpha^J = \bar{p} = \frac{1}{N}\sum_{i=1}^N p_i.
$$

#### Jackknife standard error

The jackknife variance estimate for $\alpha^J$ is based on the dispersion of the pseudo-values:

$$
\widehat{\mathrm{Var}}(\alpha^J)=\frac{1}{N(N-1)}\sum_{i=1}^N (p_i-\bar{p})^2.
$$

Hence the jackknife standard error is

$$
\sigma_\alpha
= \sqrt{\frac{1}{N(N-1)}\sum_{i=1}^N (p_i-\bar{p})^2}
= \sqrt{\frac{1}{N(N-1)}\sum_{i=1}^N \Big[N\alpha_N - \alpha^J - (N-1)\alpha_i^*\Big]^2}.
$$
